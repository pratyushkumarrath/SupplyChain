```python
drop_cols = ['Customer Email',
    'Customer Password',
    'Customer Fname',
    'Customer Lname',
    'Customer Street',
    'Customer Id',
    'Order Customer Id',
    'Order Id',
    'Order Item Id',
    'Product Image',
    'Days for shipping (real)',
    'Delivery Status'
    'Category Id',
    'Department Id',
    'Product Card Id',
    'Order Item Cardprod Id',
    'Product Status',
    'Customer City',
    'Order City']

df = df.drop(columns=drop_cols, errors='ignore')
```

```python
df.nunique()
```

```python
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'])
```

```python
df['order_year'] = df['order date (DateOrders)'].dt.year
df['order_month'] = df['order date (DateOrders)'].dt.month
df['order_day'] = df['order date (DateOrders)'].dt.day
df['order_weekday'] = df['order date (DateOrders)'].dt.weekday
df['is_weekend'] = df['order_weekday'].isin([5, 6]).astype(int)
df['week_of_year'] = df['order date (DateOrders)'].dt.isocalendar().week
df['quarter'] = df['order date (DateOrders)'].dt.quarter
df['is_month_start'] = df['order date (DateOrders)'].dt.is_month_start.astype(int)
df['is_month_end'] = df['order date (DateOrders)'].dt.is_month_end.astype(int)
```

```python
daily_orders = df.groupby(
    df['order date (DateOrders)'].dt.date
).size()

df['daily_order_volume'] = (
    df['order date (DateOrders)']
    .dt.date
    .map(daily_orders)
)
```

```python
region_daily_orders = df.groupby(
    [
        'Order Region',
        df['order date (DateOrders)'].dt.date
    ]
).size()

df['region_daily_load'] = list(zip(
    df['Order Region'],
    df['order date (DateOrders)'].dt.date
))

df['region_daily_load'] = (
    df['region_daily_load']
    .map(region_daily_orders)
)
```

```python
df['processing_time_days'] = (
    df['shipping date (DateOrders)']
    - df['order date (DateOrders)']
).dt.days
```

```python
product_processing_avg = df.groupby(
    ['Product Name', 'Order Region']
)['processing_time_days'].mean()
df['avg_product_processing_time'] = df[
    'Product Name'
].map(product_processing_avg)
```

```python
delivery_status_kpi = (
    df['Delivery Status']
    .value_counts(normalize=True) * 100
)

delivery_status_kpi
delivery_status_kpi.plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(7,7)
)

plt.title('Delivery Status Distribution')
plt.ylabel('')
plt.show()
```

```python
sns.countplot(x='Late_delivery_risk', data=df)

plt.title('Late Delivery Risk Distribution')
plt.show()
```

```python
holiday_dates = ['2015-01-01',
    '2015-07-04',
    '2015-09-07',
    '2015-11-26',
    '2015-11-27',
    '2015-11-30',
    '2015-12-25',
    '2016-01-01',
    '2016-07-04',
    '2016-09-05',
    '2016-11-24',
    '2016-11-25',
    '2016-11-28',
    '2016-12-25',
    '2017-01-01',
    '2017-07-04',
    '2017-09-04',
    '2017-11-23',
    '2017-11-24',
    '2017-11-27',
    '2017-12-25',
    '2018-01-01',
    '2018-07-04',
    '2018-09-03',
    '2018-11-22',
    '2018-11-23',
    '2018-11-26',
    '2018-12-25']
```

```python
holiday_dates = pd.to_datetime(holiday_dates)
```

```python
df['day_type'] = 'Weekday'

df.loc[
    df['order date (DateOrders)'].dt.weekday >= 5,
    'day_type'
] = 'Weekend'
holiday_dates = pd.to_datetime(holiday_dates)
df.loc[
    df['order date (DateOrders)'].dt.normalize().isin(holiday_dates),
    'day_type'
] = 'Holiday'
delay_by_day_type = df.groupby(
    'day_type'
)['Late_delivery_risk'].mean() * 100
delay_by_day_type = df.groupby(
    'day_type'
)['Late_delivery_risk'].mean() * 100
```

```python
plt.figure(figsize=(8,5))

delay_by_day_type.plot(
    kind='bar'
)

plt.title('Late Delivery Percentage by Day Type')

plt.ylabel('Late Delivery %')

plt.xticks(rotation=0)

plt.show()
```

```python
shipping_risk = df.groupby(
    'Shipping Mode'
)['Late_delivery_risk'].mean() * 100

shipping_risk.sort_values(ascending=False)
shipping_risk.sort_values().plot(
    kind='barh',
    figsize=(8,5)
)

plt.title('Delay Risk by Shipping Mode')
plt.xlabel('Delay Risk %')
plt.show()
```

```python
region_risk = df.groupby(
    ['Order Region', 'Shipping Mode']
)['Late_delivery_risk'].mean().unstack()

plt.figure(figsize=(12,6))

sns.heatmap(
    region_risk,
    annot=True,
    cmap='Reds'
)

plt.title('Regional Delay Risk Heatmap')

plt.show()
```

```python
# Monthly total orders
monthly_orders = df.groupby(
    df['order date (DateOrders)'].dt.to_period('M')
).size()

# Monthly late deliveries
monthly_late = df.groupby(
    df['order date (DateOrders)'].dt.to_period('M')
)['Late_delivery_risk'].sum()

# Monthly late delivery percentage
monthly_late_percentage = (
    monthly_late / monthly_orders
) * 100

plt.figure(figsize=(14,6))

plt.plot(
    monthly_late_percentage.index.astype(str),
    monthly_late_percentage.values,
    marker='o'
)

plt.title('Monthly Late Delivery Percentage')

plt.xlabel('Month')
plt.ylabel('Late Delivery %')

plt.xticks(rotation=45)

plt.grid(True)

plt.tight_layout()

plt.show()
```

```python
processing_time = df.groupby(
    'Category Name'
)['processing_time_days'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))

processing_time.plot(kind='bar')

plt.title('Average Processing Time by Product Category')
plt.ylabel('Processing Time (Days)')

plt.xticks(rotation=45)

plt.show()
```
